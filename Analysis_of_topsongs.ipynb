```python
#  loading dataset
dataset_path = 'Song.csv'
data = pd.read_csv(dataset_path)
```

```python
# histogram for 'Sales'
def plot_histogram(data, column, bins=10, title="Histogram", xlabel="Value", ylabel="Frequency"):
    plt.figure(figsize=(10, 6))
    sns.histplot(data[column], bins=bins, kde=True)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
# scatter graph for 'Streams' vs. 'Downloads'
def plot_scatter(data, x_column, y_column, title="Scatter Graph", xlabel="X-axis", ylabel="Y-axis"):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=data, x=x_column, y=y_column)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_heatmap(corr_matrix, title="Heatmap of Correlation Matrix"):
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title(title)
    plt.show()
```

```python
plot_histogram(data, 'Sales', bins=20, title="Histogram of Sales", xlabel="Sales (million units)", ylabel="Frequency")
```

```python
plot_scatter(data, 'Streams', 'Downloads', title="Streams vs. Downloads", xlabel="Streams (million units)", ylabel="Downloads (million units)")
```

```python
# Calculate the correlation matrix for your data
corr_matrix = data.corr(numeric_only=True)

# Call the function with the correlation matrix
plot_heatmap(corr_matrix, title="Heatmap of Correlation Matrix")
```
